```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#plt.style.use(["slide_mode","dark_mode"])
plt.style.use(["dark_mode","slide_mode"])
#plt.style.use("dark_mode")
#lt.style.use("default")
ddf = 'default'
```

```python
x = np.random.normal(5,3,1000)
_ = plt.hist(x)
```

```python
import pgsahist as ph
```

```python
hre = ph.hist.Hist1D(x,bins=10)
hre.plot(steps=True,filled=False,ebar=True,xlabel='random value',ylabel='Count',title='Histogram Demo')
plt.savefig("../../Slides/images/demo_hist_{}.png".format(ddf),dpi=150,bbox_inches='tight')
```

```python
h = ph.hist.Hist1D(x,bins=10).normalize().fit_normal()
h.plot(xlabel="random variable",ylabel="PDF",title="Example curve fitting")
plt.savefig("../../Slides/images/demo_fit_{}.png".format(ddf),dpi=150,bbox_inches='tight')
```

```python
hh = h/3
hh.plot()
```
